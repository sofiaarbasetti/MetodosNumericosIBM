## FunciÃ³n-CÃ³digo

```python
def err_c(data_x,data_y):
    size = len(data_x);
    i=0
    sum_xy=0
    sum_y=0
    sum_x=0
    sum_sqare_x=0
    average_x=0
    average_y=0
    while i<size:
        sum_xy+=data_x[i]*data_y[i]
        sum_y+=data_y[i]
        sum_x+=data_x[i]
        sum_sqare_x+=data_x[i]*data_x[i]
        i+=1
    average_x=sum_x/size
    average_y=sum_y/size
    return_k=(size*sum_xy-sum_x*sum_y)/(size*sum_sqare_x-sum_x*sum_x)
    return_b=average_y-average_x*return_k
    # Plot 
    x = np.linspace(0,20,100) 
    recta = return_k*x + return_b
    fig, ax = plt.subplots()
    ax.scatter(puntos_x, puntos_y);
    ax.plot(x,recta,label='Recta')
    font = font_manager.FontProperties(weight='medium', style='normal', size=16)
    ax.legend(prop=font)
    
    return print('a1: {0} - a0: {1}'.format(return_k,return_b))


data_x = df['x']
data_y = df['y']
err_c(data_x,data_y)
```
