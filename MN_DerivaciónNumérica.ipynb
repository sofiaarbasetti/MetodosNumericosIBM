# Metodo: Derivación Numérica

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.linear_model import LinearRegression 
regresion_lineal = LinearRegression()
```

## Ejemplo

```python
def funcion_ejemplo(x):
    return (x**2)
```

```python
puntos_x =  np.arange(0,20,0.5)
puntos_y = funcion_ejemplo(puntos_x)

fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(14,8))
ax.plot(puntos_x,puntos_y,label='f(x)');

# puntos seleccionados
puntos_selecionados_x = np.array([7.5,17.5])
puntos_selecionados_y = funcion_ejemplo(puntos_selecionados_x)
##--- Puntos de referencia
ax.scatter(puntos_selecionados_x, puntos_selecionados_y, s=46, marker='o', c='k')
ax.vlines(x = puntos_selecionados_x, ymin = [0,0], ymax = puntos_selecionados_y,
          linestyle='--')
ax.hlines(y = puntos_selecionados_y, xmin = 0, xmax = puntos_selecionados_x, linestyle='--')

##--- Recta entre los 2 puntos

#ax.plot(puntos_selecionados_x,puntos_selecionados_y)
regresion_lineal.fit(puntos_selecionados_x.reshape(-1,1), puntos_selecionados_y) 
ax.plot(puntos_x,(regresion_lineal.coef_*puntos_x+regresion_lineal.intercept_),label="recta")

#Axis
ax.axis([0, 20, 0 , 400])


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend();
```

## Definición de Metodos

```python
import sympy as sp
x = sp.symbols('x')
f1 = 4*x**2+3
f2 = 5*x**3+3*x
```

```python
f1.subs(x,2) # Evalua la funcion en un punto con sympy
```

```python
def adelante(f,xi,h):
    '''derivación numérica hacia adelante, 
    f: es la función utilizando sympy, 
    xi: el punto donde evaluar y 
    h: es la distancia entre xi+1 - xi'''
    result = (f.subs(x,xi+h)-f.subs(x,xi))/h

    return result

def atras(f,xi,h):
    '''derivación numérica hacia adelante, 
    f: es la función utilizando sympy, 
    xi: el punto donde evaluar y 
    h: es la distancia entre xi - xi-1'''
    result = (f.subs(x,xi)-f.subs(x,xi-h))/h

    return result

def centrado(f,xi,h):
    '''derivación numérica hacia adelante, 
    f: es la función utilizando sympy, 
    xi: el punto donde evaluar y 
    h: es la distancia entre xi - xi-1 o xi+h-xi'''
    result = (f.subs(x,xi+h)-f.subs(x,xi-h))/(2*h)

    return result

def real(f,xi):
    '''derivación numérica real, 
    f: es la función utilizando sympy, 
    xi: el punto donde evaluar y '''
    result = (f.diff()).subs(x,xi)

    return result
```

### Ejemplo F1

```python
adelante(f1,2,1)
```

```python
atras(f1,2,1)
```

```python
centrado(f1,2,1)
```

```python
real(f1,2)
```

#### Array de h

```python
h1 = [2,1,0.1,0.001]
for h in h1:
    print('Para h={0}, hacia adelante da: {1}'.format(h,adelante(f1,2,h)))
    print('Para h={0}, hacia atras da: {1}'.format(h,atras(f1,2,h)))
    print('Para h={0}, centrado da: {1}'.format(h,centrado(f1,2,h)))
    print('Real: {}'.format(real(f1,2)))
    print('\n')
```

```python
e_adelante_f1 = []
e_atras_f1 = []
e_centrado_f1 = []

for h in h1:
  e_adelante_f1.append(abs(real(f1,2)-adelante(f1,2,h))) 
  e_atras_f1.append(abs(real(f1,2)-atras(f1,2,h))) 
  e_centrado_f1.append(abs(real(f1,2)-centrado(f1,2,h))) 

#TABLA
errores_f1 = {
    'adelante' : e_adelante_f1,
    'atras' :    e_atras_f1,
    'centrado' : e_centrado_f1,
    }

errores_f1 = pd.DataFrame(errores_f1)
errores_f1 = errores_f1.astype(float) 

errores.head()

#PLOTS
errores_f1['adelante'].plot()
errores_f1['atras'].plot() 
errores_f1['centrado'].plot()
```
